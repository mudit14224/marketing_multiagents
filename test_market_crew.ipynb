```python
search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()
```

```python
# Agents 
marketing_agent = Agent(
    role="Sales and Marketing Head",
    goal="Find out all information about a healthcare provider",
    tools=[search_tool, scrape_tool],
    backstory = """You work for a digital healthcare company called as Healee. You are the head od the sales and 
    marketing. You are looking for potential healthcare provider clients similar to {client} that your company can
    pitch your platform to. """,
    verbose=True
)
```

```python
scrape_task = Task(description="""Look into {client} website and scrape and extract all the relevant information about the company.""",
            agent=marketing_agent,
            expected_output="""All the details about the {client} in a concise summary that can be used to build search queries to 
            find similar companies.""")

search_task = Task(description="""Find companies similar to the {client} that the marketing team can pitch the Healee platform to.""",
                   agent=marketing_agent,
                   expected_output="A list of potential companies along with their website links and contact information.")
```

```python
crew = Crew(
    agents=[marketing_agent],
    tasks=[scrape_task, search_task],
    verbose=True
)

inputs = {
    "client": "Icon Health"
}
result = crew.kickoff(inputs=inputs)
```
